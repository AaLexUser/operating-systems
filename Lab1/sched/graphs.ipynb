```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./csv/schedpolicy-pidstat.csv', delimiter=";")
df.head()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Time'], df['policy'])
plt.xlabel('Time')
plt.ylabel('Policy')
plt.grid(True)
plt.minorticks_on()
plt.grid(which='minor', linestyle='-', color='#999999', alpha=0.2)
plt.title('Policy over Time')
plt.savefig('./image/schedpolicy-pidstat.png')
plt.show()
```
