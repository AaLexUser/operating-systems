```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cpu-stat.csv', delimiter=';')
df.head()
```

```python
df.isna().sum()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df['%CPU'] = df['%CPU'].str.replace(',', '.').astype(float)
df
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Time'], df['%CPU'])
plt.xlabel('Time')
plt.ylabel('CPU %')
plt.grid(True)
plt.minorticks_on()
plt.grid(which='minor', linestyle='-', color='#999999', alpha=0.2)
plt.title('CPU % over Time')
plt.savefig('./image/cpu_usage.png')
plt.show()
```

```python
df = pd.read_csv('./csv/nice-testing.csv', delimiter=';')
df['sys+user'] = df['usr_time'] + df['sys_time']
df.head()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(20,15))

axs[0,0].plot(df['nice'], df['bogo_ops'])
axs[0,0].set_xlabel('nice')
axs[0,0].set_ylabel('Bogo ops')
axs[0,0].set_title('Bogo ops over nice')

axs[0,1].plot(df['nice'], df['bogo_ops/s_real'])
axs[0,1].set_xlabel('nice')
axs[0,1].set_ylabel('Bogo ops/s (real time)')
axs[0,1].set_title('Bogo ops/s over nice')

axs[1,0].plot(df['nice'], df['bogo_ops/s_user+sys'])
axs[1,0].set_xlabel('nice')
axs[1,0].set_ylabel('Bogo ops/s (sys+user)')
axs[1,0].set_title('Bogo ops/s over nice')

axs[1,1].plot(df['nice'], df['CPU_used_per'])
axs[1,1].set_xlabel('nice')
axs[1,1].set_ylabel('CPU used per')
axs[1,1].set_title('CPU used over nice')

axs[2,0].plot(df['nice'], df['real_time'])
axs[2,0].set_xlabel('nice')
axs[2,0].set_ylabel('Real Time')
axs[2,0].set_title('Real Time over L1cache sets')

axs[2,1].plot(df['nice'], df['sys+user'])
axs[2,1].set_xlabel('nice')
axs[2,1].set_ylabel('Sys + User Time')
axs[2,1].set_title('Sys + User Time over nice')

for i in range(3):
    for j in range(2):
        axs[i,j].grid(True)
        axs[i,j].minorticks_on()
        axs[i,j].grid(which='minor', linestyle='-', color='#999999', alpha=0.2)

plt.savefig('./image/nice-testing.png')
plt.show()
```

```python
df = pd.read_csv('./csv/sched-tesing.csv', delimiter=";")
df['sys+user'] = df['usr_time'] + df['sys_time']
df.head()
```

```python
x_column = 'sched'
fig, axs = plt.subplots(3, 2, figsize=(20,15))

axs[0,0].plot(df[x_column], df['bogo_ops'])
axs[0,0].set_xlabel(x_column)
axs[0,0].set_ylabel('Bogo ops')
axs[0,0].set_title('Bogo ops over ' + x_column)

axs[0,1].plot(df[x_column], df['bogo_ops/s_real'])
axs[0,1].set_xlabel(x_column)
axs[0,1].set_ylabel('Bogo ops/s (real time)')
axs[0,1].set_title('Bogo ops/s over ' + x_column)

axs[1,0].plot(df[x_column], df['bogo_ops/s_user+sys'])
axs[1,0].set_xlabel(x_column)
axs[1,0].set_ylabel('Bogo ops/s (sys+user)')
axs[1,0].set_title('Bogo ops/s over ' + x_column)

axs[1,1].plot(df[x_column], df['CPU_used_per'])
axs[1,1].set_xlabel(x_column)
axs[1,1].set_ylabel('CPU used per')
axs[1,1].set_title('CPU used over ' + x_column)

axs[2,0].plot(df[x_column], df['real_time'])
axs[2,0].set_xlabel(x_column)
axs[2,0].set_ylabel('Real Time')
axs[2,0].set_title('Real Time over ' + x_column)

axs[2,1].plot(df[x_column], df['sys+user'])
axs[2,1].set_xlabel(x_column)
axs[2,1].set_ylabel('Sys + User Time')
axs[2,1].set_title('Sys + User Time over ' + x_column)

for i in range(3):
    for j in range(2):
        axs[i,j].grid(True)
        axs[i,j].minorticks_on()
        axs[i,j].grid(which='minor', linestyle='-', color='#999999', alpha=0.2)

plt.savefig('./image/sched-tesing.png')
plt.show()
```
