```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cpu-stat.csv', delimiter=';')
df.head()
```

```python
df.isna().sum()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df['%CPU'] = df['%CPU'].str.replace(',', '.').astype(float)
df
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Time'], df['%CPU'])
plt.xlabel('Time')
plt.ylabel('CPU %')
plt.grid(True)
plt.minorticks_on()
plt.grid(which='minor', linestyle='-', color='#999999', alpha=0.2)
plt.title('CPU % over Time')
plt.savefig('./image/cpu_usage.png')
plt.show()
```
